## 1-2. 이상치 및 결측치 찾기

컬럼별로 돌아가며 이상한 애들을 찾는다

```python
for col in test2017.columns[129:140]: #여기서 연도별로 슬라이스 해줘가며 컬럼별로 유니크값을 본다
# for col in test2018.columns[129:140]:
# for col in test2019.columns[129:140]:
# for col in test2020.columns[129:140]:
  print(col)
  L = test2017[col].unique()
  # L = test2018[col].unique()
  # L = test2019[col].unique()
  # L = test2020[col].unique()
  try:
    L.sort() #값들이 순서대로 나옴 -> 가독성 좋게보려고
    if len(L) < 10:
      print(L)
    else:
      for i in range(10,len(L),10):
        print(L[i-10:i])
      print(L[i:])
      print()
    print()
  except:
    print(f'{col} 이상함 발견됨!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!') #이건 sort안되는걸 방지
    print()

# EX) 예를 들어 보기가 1~5 번부터 인데 여기서 나온 값이 [1,2,3,4,5,8, 2900] 이면 이상한 값이 껴있다고 유추가능
```

## 1-3. Column 순회하면서 발견한점 정리
- 객관식답에 주관식 답이 들어가 있거나
- 주관식 답에 객관식 답이 들어가 있는걸 심심치 않게 볼 수 있다
- 이에따라 몇몇 Row 값들이 어느 부분에서 밀려 적힌곳이 있을 거라고 유추가능
- 밀려적힌 값들을 우선 찾는게 중요하다고 판단됨


Train Data와 같은 방식으로 뒤에 Nan값이 있는 Row들을 찾아 밀려주는 값들을 원래 위치로 맞추는 작업을 한다

### 2017년 : 딱히 문제되는거 없음

```python
c_list = test2017.columns.tolist()
nan2017 = []

for i in c_list:
    nan2017.append([i,list(test2017[test2017[i].isnull()].index)])
nan2017 #2017년에서는 발견되지 않음
```
