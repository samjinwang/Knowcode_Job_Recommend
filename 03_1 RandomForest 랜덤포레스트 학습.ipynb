```python
from bayes_opt import BayesianOptimization
```

```python
def rf_bo(max_depth, n_estimators, random_state):

  rf_params = {
              'max_depth' : int(round(max_depth)),
              'n_estimators' : int(round(n_estimators)), 
              'random_state' : int(round(random_state)
              }

  rf = RFC(**rf_params)
  X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size = 0.2, stratify = y)

  rf.fit(X_train,y_train)
  score = f1_score(y_valid, rf.predict(X_valid), average = 'macro')

  return score
```

## 3.1 2017년

학습범위설정

```python
rf_parameter_bounds = {
                      'max_depth' : (20,35),
                      'n_estimators' : (175,350),   
                      'random_state' : (10,35)
                      }
```

```python
BO_rf_2017 = BayesianOptimization(f = rf_bo, pbounds = rf_parameter_bounds)
```

학습

하이퍼 파라미터 숫자들을 더한만큼 학습됨

```python
BO_rf_2017.maximize(init_points = 5, n_iter = 30)
```
