# 2. 연도별로 학습을 통해 찾은 파라미터를 적용해 KNOWCODE를 테스트에 추가한다
- Voting Classifier로 두가지 모델의 성능을 통합한 제출 파일을 만든다

2017년

```python
et = ExtraTreesClassifier(max_depth = 30, n_estimators = 300, random_state = 30)
rf = RandomForestClassifier(max_depth = 25, n_estimators = 350)
```

```python
model = VotingClassifier(estimators=[('rf',rf),('et',et)],voting = 'soft')
```

```python
train_x = meta2017.drop(['knowcode'], axis = 1)
train_y = meta2017['knowcode']
test_x = test2017
```

```python
model.fit(train_x, train_y)
pred2017 = model.predict(test_x)
```
